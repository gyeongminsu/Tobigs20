### 1. Normal Equation

```python
import numpy as np
from numpy.linalg import inv
```

```python
def estimate_beta(X, y):
    designX = np.c_[np.ones((X.shape[0], 1)), X]
    beta_hat = inv(designX.T@designX)@designX.T@y

    return beta_hat
```

```python
beta_hat = estimate_beta(X, y)
beta_hat
```

### 2. MSE

```python
designX = np.c_[np.ones((X.shape[0], 1)), X]
designX.dot(beta_hat)
```

```python
def MSE(X, y, beta_hat):
    designX = np.c_[np.ones((X.shape[0], 1)), X]
    y_pred = designX.dot(beta_hat)
    mse = np.linalg.norm((y - y_pred)**2) / len(y)

    return mse
```

```python
MSE(X, y, estimate_beta(X, y))
```
